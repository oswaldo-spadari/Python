```python
import pandas as pd
import yfinance as yf
from plyer import notification


# Defina as configurações das médias móveis
short_window = 9  # Período da EMA curto
long_window = 21  # Período da EMA longo

# Defina a lista de símbolos das ações a serem monitoradas 
stock_symbols = ['RRRP3.SA', 'ALSO3.SA', 'ALPA4.SA', 'ABEV3.SA', 'ARZZ3.SA', 'ASAI3.SA', 'AURE3.SA', 'AZUL4.SA', 'B3SA3.SA', 'BBSE3.SA', 'BBDC3.SA', 'BBDC4.SA', 'BRAP4.SA', 'BBAS3.SA', 'BRKM5.SA', 'BRFS3.SA', 'BPAC11.SA', 'CRFB3.SA', 'CBAV3.SA', 'CCRO3.SA', 'CMIG4.SA', 'CIEL3.SA', 'COGN3.SA', 'CPLE6.SA', 'CSAN3.SA', 'CPFE3.SA', 'CMIN3.SA', 'CVCB3.SA', 'CYRE3.SA', 'DXCO3.SA', 'DIRR3.SA', 'ECOR3.SA', 'ELET3.SA', 'ELET6.SA', 'EMBR3.SA', 'ENGI11.SA', 'ENEV3.SA', 'EGIE3.SA', 'EQTL3.SA', 'EZTC3.SA', 'FLRY3.SA', 'GGBR4.SA', 'GOAU4.SA', 'GOLL4.SA', 'GMAT3.SA', 'NTCO3.SA', 'SOMA3.SA', 'HAPV3.SA', 'HYPE3.SA', 'IGTI11.SA', 'IRBR3.SA', 'ITSA4.SA', 'ITUB4.SA', 'JBSS3.SA', 'KLBN11.SA', 'RENT3.SA', 'LWSA3.SA', 'LREN3.SA', 'MDIA3.SA', 'MGLU3.SA', 'MRFG3.SA', 'CASH3.SA', 'BEEF3.SA', 'MOVI3.SA', 'MRVE3.SA', 'MULT3.SA', 'PCAR3.SA', 'PETR3.SA', 'PETR4.SA', 'RECV3.SA', 'PRIO3.SA', 'PETZ3.SA', 'PSSA3.SA', 'QUAL3.SA', 'RADL3.SA', 'RAIZ4.SA', 'RDOR3.SA', 'RAIL3.SA', 'SBSP3.SA', 'SANB11.SA', 'STBP3.SA', 'SMTO3.SA', 'CSNA3.SA', 'SIMH3.SA', 'SLCE3.SA', 'SUZB3.SA', 'TAEE11.SA', 'VIVT3.SA', 'TIMS3.SA', 'TOTS3.SA', 'TRPL4.SA', 'UGPA3.SA', 'USIM5.SA', 'VALE3.SA', 'VAMO3.SA', 'VIIA3.SA', 'VBBR3.SA', 'WEGE3.SA', 'YDUQ3.SA']
#acima estão as 100 mais negociadas da B3

# Defina a data de início para a obtenção dos dados das ações
start_date = '2023-01-15'

# Lista para armazenar as ações com sinal gerado
actions_with_signals = []

# Função para verificar os sinais e enviar notificações
def check_signals():
    for stock_symbol in stock_symbols:
        data = yf.download(stock_symbol, start=start_date, interval='1d')
        data['EMA_short'] = data['Close'].ewm(span=short_window, adjust=False).mean()
        data['EMA_long'] = data['Close'].ewm(span=long_window, adjust=False).mean()

        data['Signal'] = 0
        data.loc[data['EMA_short'] > data['EMA_long'], 'Signal'] = 1

        last_signal = data.iloc[-2]['Signal'] if len(data) >= 2 else 0
        current_signal = data.iloc[-1]['Signal']

        if current_signal != last_signal:
            if current_signal == 1:
                signal_type = 'Compra'
            else:
                signal_type = 'Venda'
            
            title = f"Sinal de {signal_type} para {stock_symbol}"
            message = "Um novo sinal foi gerado."
            notification.notify(title=title, message=message)
            actions_with_signals.append(f"{stock_symbol} ({signal_type})")
    
    if actions_with_signals:
        print("Ações com sinais gerados:", ", ".join(actions_with_signals))

        
if __name__ == '__main__':
    check_signals()
```
