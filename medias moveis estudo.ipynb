```python
import pandas as pd
import yfinance as yf
from plyer import notification
import time

# Defina as configurações das médias móveis
short_window = 9  # Período da EMA curto
long_window = 21  # Período da EMA longo

# Defina a lista de símbolos das ações a serem monitoradas
stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN']  # Exemplos de ações

# Defina a data de início para a obtenção dos dados das ações
start_date = '2022-01-01'

# Função para verificar os sinais e enviar notificações
def check_signals():
    for stock_symbol in stock_symbols:
        data = yf.download(stock_symbol, start=start_date)
        data['EMA_short'] = data['Close'].ewm(span=short_window, adjust=False).mean()
        data['EMA_long'] = data['Close'].ewm(span=long_window, adjust=False).mean()

        data['Signal'] = 0
        data.loc[data['EMA_short'] > data['EMA_long'], 'Signal'] = 1

        last_signal = data.iloc[-2]['Signal'] if len(data) >= 2 else 0
        current_signal = data.iloc[-1]['Signal']

        if current_signal != last_signal:
            title = f"Sinal de {'Compra' if current_signal == 1 else 'Venda'} para {stock_symbol}"
            message = "Um novo sinal foi gerado."
            notification.notify(title=title, message=message)

def main():
    # Agende a execução a cada X minutos (ajuste conforme necessário)
    interval_minutes = 5
    while True:
        check_signals()
        time.sleep(interval_minutes * 60)  # Converter minutos para segundos

if __name__ == '__main__':
    main()
```
